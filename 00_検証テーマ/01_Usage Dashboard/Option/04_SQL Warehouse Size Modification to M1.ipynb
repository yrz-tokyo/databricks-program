```python
import requests

workspace = ""
token = ""
warehouse_id = ""

url = f"{workspace}/api/2.0/sql/warehouses/{warehouse_id}/edit"

payload = {
    "min_num_clusters": 1,
    "max_num_clusters": 1,
    "cluster_size": "Medium"
}

headers = {
    "Authorization": f"Bearer {token}",
    "Content-Type": "application/json"
}

resp = requests.post(url, json=payload, headers=headers)
print(resp.status_code, resp.text)
```
