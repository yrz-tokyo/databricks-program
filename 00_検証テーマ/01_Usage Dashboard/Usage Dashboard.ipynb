# 予算管理（Public Preview）の紹介

・予算管理の機能で、アカウント全体、或いは特定ワークスペース、特定プロジェクト（タグ利用）の使用量をトラックし、設定された閾値を超えるとメール通知もできます。<br>
・予算の単位はUSDです。<br>
・毎月の累積支出が選択したしきい値を超えた場合、予算使用後24時間以内にアラートが送信されます。<br>
