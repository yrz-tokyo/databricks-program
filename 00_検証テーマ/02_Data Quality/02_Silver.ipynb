```python
%sql
CREATE OR REPLACE TABLE dataquality.dq_demo.silver_bev_events
USING DELTA
AS
SELECT
  CAST(event_ts AS TIMESTAMP)          AS event_ts,
  ingest_hour,

  txn_id,
  plant_id,

  -- NULLはそのまま残す（重要）
  warehouse_id,
  store_id,

  channel,
  sku_id,

  package_ml,
  qty,
  unit_price_yen,
  sales_yen,

  -- 整合性確認用（直さない）
  qty * unit_price_yen AS expected_sales_yen,

  delivery_lead_hours,
  promo_flag,
  quality_flag,

  ingested_at
FROM dataquality.dq_demo.bronze_bev_events;
```
