```python
%sql
CREATE OR REPLACE TABLE dataquality.dq_demo.gold_sales_hourly
USING DELTA
AS
SELECT
  ingest_hour,
  channel,
  sku_id,

  COUNT(*)                AS txn_cnt,
  SUM(qty)                AS total_qty,
  SUM(sales_yen)          AS total_sales_yen,

  -- あえて品質が影響する形
  COUNT(DISTINCT txn_id)  AS distinct_txn_cnt
FROM dataquality.dq_demo.silver_bev_events
GROUP BY ingest_hour, channel, sku_id;
```
